```python
df = df.drop_duplicates()
print(df)
```

```python
total_sales = df["Sales_Amount"].sum()

print("Total sales:",total_sales)

```

```python
prodect_sales = df.groupby("sales_Rep")["Sales_Amount"].sum()

print(prodect_sales)
```

```python
prodect_sales.plot(kind="bar")

plt.title("product sales Analysis")

plt.xlabel("products")

plt.show()
```
