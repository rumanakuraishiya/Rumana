```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

upload=files.upload()

# Get the first uploaded filename (assuming only one file is expected)
file_name = list(upload.keys())[0]
df=pd.read_excel(file_name)

print(df)
```

DASHBOARD

```python
df = pd.read_excel(file_name)

df['TOTAL PRICE'] = df['TOTAL PRICE'].astype(str).str.replace('$', '', regex=False).astype(float)

print("TOTAL ORDERS:", len(df))
print("TOTAL SALES:$ ", df["TOTAL PRICE"].sum())

df["FOOD ITEM"].value_counts().plot(kind="bar")
plt.title("FOOD ORDERS")
plt.show()

df["LOCATION"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("ORDERS BY LOCATION")
plt.show()

```
